```python
import pandas as pd

# dataset used is an open data set which comes from kaggle
# source: https://www.kaggle.com/doaaalsenani/usa-cers-dataset/data#

filepath = '/Users/tianyiwei/Desktop/wty/'
df_car = pd.read_csv(filepath+'USA_cars_datasets.csv')
```

```python
df_car = df_car.iloc[:,1:]
df_car
```

```python
df_car.info()  # columns, dtype, na
```

```python
df_car.describe()  # count/mean/min/max  #stats
```

### Sort Values

```python
# list top5 price
df_car.sort_values(by='price',ascending=False).head()

# sort by multiple columns
df_car.sort_values(by=['price','year'],ascending=[False,False]).head()  # same price with latest year
```

### Subset Columns

```python
# subset part of the columns
df_car_pbm = df_car[['price','brand','model']]
df_car_pbm.head()
```

### Subset Rows

```python
# multiple conditions filter
df_car[(df_car['price']<=22000)&(df_car['brand']=='chevrolet')&(df_car['year']>=2019)].head()

# put true flase in [] to filter: df_car[a series of True/False]
```

```python
# Subsetting rows by categorical variables
my_favorite_brand = ['chevrolet','cadillac','audi']
condition = df_car['brand'].isin(my_favorite_brand)

df_car[df_car['brand'].isin(my_favorite_brand)]  # subset
```

### Adding New Columns

```python
df_car['price_in_thousand'] = df_car['price']/1000
df_car.head()
```
