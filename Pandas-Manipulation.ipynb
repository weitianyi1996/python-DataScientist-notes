```python
# four most common types of data manipulation: 
# sorting rows, 
# subsetting columns, 
# subsetting rows, 
# adding new columns
```

```python
import pandas as pd

# dataset used is an open data set which comes from kaggle
# source: https://www.kaggle.com/doaaalsenani/usa-cers-dataset/data

filepath = '/Users/tianyiwei/Desktop/wty/'
df_car = pd.read_csv(filepath+'USA_cars_datasets.csv')
```

```python
df_car = df_car.iloc[:,1:]
df_car
```

```python
df_car.info()  # columns, dtype, na
```

```python
df_car.describe()  # count/mean/min/max  #stats
```

### Sort Values

```python
# list top5 price
df_car.sort_values(by='price',ascending=False).head()

# sort by multiple columns
df_car.sort_values(by=['price','year'],ascending=[False,False]).head()  # same price with latest year
```

### Subset Columns

```python
# subset part of the columns
df_car_pbm = df_car[['price','brand','model']]
df_car_pbm.head()
```

### Subset Rows

```python
# multiple conditions filter
df_car[(df_car['price']<=22000)&(df_car['brand']=='chevrolet')&(df_car['year']>=2019)].head()

# put true flase in [] to filter: df_car[a series of True/False]
```

```python
# Subsetting rows by categorical variables
my_favorite_brand = ['chevrolet','cadillac','audi']
condition = df_car['brand'].isin(my_favorite_brand)

df_car[df_car['brand'].isin(my_favorite_brand)]  # subset
```

### Adding New Columns

```python
df_car['price_in_thousand'] = df_car['price']/1000
df_car.head()
```

### Aggregating

```python
df_car['price'].max()
```

```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

df_car['price'].agg(iqr)
```

```python
df_car['price'].cumsum()  #cumulative sum-max-aggregating function can be a series of data instead of 1 value
```

```python
# drop duplicates
df_car.drop_duplicates(subset=['brand','model','year'])
```

```python
# count categorical variables
df_car['brand'].value_counts(sort=True, normalize=True)
```

```python
# group by aggregation
df_car.groupby(by=['brand'])['price'].mean()  # df_car.groupby(by=['brand','year'])['price'].mean()
```

```python
# group by and multiple statistics
import numpy as np

df_car.groupby(by='brand')['price'].agg([np.min,np.max,np.mean])
```

```python
# pivot tables(alternative to group by)
df_car.pivot_table(index=['brand'],columns=['year'],values='price',aggfunc=[np.mean],fill_value=0)
```

### Slicing and Indexing

```python
df_car_index = df_car.set_index(['brand','model'])  #  multi-level
# df_car_index.reset_index()
df_car_index 
```

```python
df_car_index = df_car_index.sort_index()

# take subsets in both directions at once
df_car_index.loc[('acura','door'):('audi','door'), 'price':'color'] 
```

```python
# filter rows by a range of values at 1 col
df_car[(df_car['price']>=15000)&(df_car['price']<=25000)]
```
